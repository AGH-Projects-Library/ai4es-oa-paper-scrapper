```python
# Import exported article JSON

# This notebook loads the aggregated export JSON, lists processed documents, and renders section tables in a form that is easy to inspect and re-import.
```

```python
import json
from pathlib import Path
import pandas as pd

EXPORT_PATH = Path("../paper_pipeline_data/exports/processed_export_1778606250.json")


def load_export(path=EXPORT_PATH):
    path = Path(path)
    with path.open("r", encoding="utf-8") as f:
        return json.load(f)


def list_documents(data):
    for idx, doc in enumerate(data, start=1):
        print(f"{idx}. {doc.get('paper_id')} [{doc.get('source')}]")


def table_to_dataframe(table):
    header = table.get("header") or []
    rows = table.get("rows") or []
    if not header:
        return pd.DataFrame(rows)
    return pd.DataFrame(rows, columns=header)


def render_document_tables(doc):
    print("\nDocument metadata:")
    print(
        {
            k: doc.get(k)
            for k in ["paper_id", "source", "pmcid", "arxiv_id", "md_path", "html_path", "authors", "emails"]
            if k in doc
        }
    )

    sections = doc.get("sections", [])
    print(f"\nSections: {len(sections)}")

    for section in sections:
        tables = section.get("tables", [])
        if not tables:
            continue

        print(f"\n## {section.get('heading', '(untitled)')}")
        for idx, table in enumerate(tables, start=1):
            df = table_to_dataframe(table)
            print(f"\nTable {idx}")
            display(df)


data = load_export()
print(f"Loaded {len(data)} documents from {EXPORT_PATH}")
list_documents(data)
```

```python
if data:
    choice = 1
    if len(data) > 1:
        print("\nSelect a document to render:")
        list_documents(data)
        try:
            choice = int(input("Choose document number: ").strip())
        except Exception:
            choice = 1

    choice = max(1, min(choice, len(data)))
    selected = data[choice - 1]
    render_document_tables(selected)
```
